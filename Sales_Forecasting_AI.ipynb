```python
# ============================================================
# SALES FORECASTING USING AI
# ============================================================

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Step 1: Create dataset
data = {
    'date': [
        '2024-01-01','2024-01-02','2024-01-03','2024-01-04','2024-01-05',
        '2024-01-06','2024-01-07','2024-01-08','2024-01-09','2024-01-10'
    ],
    'sales': [200, 220, 250, 270, 300, 320, 350, 370, 400, 420]
}

df = pd.DataFrame(data)
df['date'] = pd.to_datetime(df['date'])

# Step 2: Feature Engineering
df['day_number'] = range(1, len(df)+1)

# Step 3: Define variables
X = df[['day_number']]
y = df['sales']

# Step 4: Train model
model = LinearRegression()
model.fit(X, y)

# Step 5: Predict next 3 days
future_days = np.array([[11], [12], [13]])
predictions = model.predict(future_days)

for i, pred in enumerate(predictions, start=11):
    print(f"Predicted Sales for Day {i}: {pred}")

# Step 6: Visualization
plt.figure(figsize=(8,5))
plt.scatter(X, y, label="Actual Sales")
plt.plot(X, model.predict(X), label="Trend Line")
plt.scatter(future_days, predictions, label="Predicted Sales")
plt.xlabel("Day Number")
plt.ylabel("Sales")
plt.title("Sales Forecasting using AI")
plt.legend()
plt.show()

# Step 7: Save Output
df.to_csv("sales_data.csv", index=False)
```

```python
# Save DataFrame to Excel file
df.to_excel("sales_data.xlsx", index=False)

print("Excel file saved successfully!")
```
